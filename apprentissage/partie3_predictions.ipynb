```python
# Chargement du Pokédex et lancement d'un combat
with open("datas/pokedex.csv", newline='') as csvfile:
    pokedex = csv.reader(csvfile)
    # On saute la première ligne (les noms de colonnes)
    next(pokedex)

    # Exemple de combat (tu peux changer les numéros)
    prediction(368, 598, pokedex)
```

```python
def prediction(numeroPokemon1, numeroPokemon2, Pokedex):
    # On récupère les infos des deux Pokémons
    pokemon1 = rechercheInformationsPokemon(numeroPokemon1, Pokedex)
    pokemon2 = rechercheInformationsPokemon(numeroPokemon2, Pokedex)

    # Chargement du modèle entraîné (RandomForest) sauvegardé dans partie3.ipynb
    modele_prediction = joblib.load('modele/modele_pokemon.mod')

    # On prépare les 7 features pour chaque Pokémon
    # (points de vie, niv_attaque, niv_defense, niv_attaque_spec,
    #  niv_defense_spec, vitesse, génération)
    prediction_Pokemon_1 = modele_prediction.predict([[
        float(pokemon1[2]),
        float(pokemon1[3]),
        float(pokemon1[4]),
        float(pokemon1[5]),
        float(pokemon1[6]),
        float(pokemon1[7]),
        float(pokemon1[8])
    ]])

    prediction_Pokemon_2 = modele_prediction.predict([[
        float(pokemon2[2]),
        float(pokemon2[3]),
        float(pokemon2[4]),
        float(pokemon2[5]),
        float(pokemon2[6]),
        float(pokemon2[7]),
        float(pokemon2[8])
    ]])

    print(f"COMBAT OPPOSANT : ({numeroPokemon1}) {pokemon1[1]} à ({numeroPokemon2}) {pokemon2[1]}")
    print(f"  {pokemon1[1]} : {prediction_Pokemon_1[0]}")
    print(f"  {pokemon2[1]} : {prediction_Pokemon_2[0]}")
    print("")

    if prediction_Pokemon_1 > prediction_Pokemon_2:
        print(pokemon1[1].upper() + " EST LE VAINQUEUR !")
    else:
        print(pokemon2[1].upper() + " EST LE VAINQUEUR !")
```

```python
def rechercheInformationsPokemon(numPokemon, Pokedex):
    """
    Cherche les infos d'un Pokémon dans le pokedex (par son NUMERO).

    Retourne une liste :
    [numero, nom, points_de_vie, niv_attaque, niv_defense,
     niv_attaque_speciale, niv_defense_speciale, vitesse, generation]
    """
    infosPokemon = []
    for pokemon in Pokedex:
        # pokemon[0] = NUMERO (dans le CSV), c'est une string -> on la convertit en int
        if int(pokemon[0]) == numPokemon:
            infosPokemon = [
                pokemon[0],  # numéro
                pokemon[1],  # nom
                pokemon[4],  # points de vie
                pokemon[5],  # niveau attaque
                pokemon[6],  # niveau défense
                pokemon[7],  # niveau attaque spéciale
                pokemon[8],  # niveau défense spéciale
                pokemon[9],  # vitesse
                pokemon[10]  # génération
            ]
            break

    return infosPokemon
```
